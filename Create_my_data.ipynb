```python
import numpy as np
import pandas as pd
```

### My data

```python
columns_name = np.array([
        'mod_zurix',             # 1. Код для класса (модифицированный класс Цюриха) (A, B, C, D, E, F, H)
        'largest_spot',          # 2. Код для наибольшего размера пятна (X, R, S, A, H, K)
        'spot_distribution',     # 3. Код для пятна распределение (X, O, I, C)
        'activity',              # 4. Активность (1 = уменьшенная, 2 = неизменная)
        'evolution',             # 5. Эволюция (1 = распад, 2 = отсутствие роста, 3 = рост)
        'activity_last_24',      # 6. Предыдущий 24-часовой код активности вспышки (1 = ничто так велико, как M1, 2 = один M1, 3 = больше активности, чем один M1)
        'historically_complex',  # 7. Исторически сложный (1 = Да, 2 = Нет)
        'suns_disk',             # 8. Стал ли исторически сложным регион на этом проходе через солнечный диск (1 = да, 2 = нет)
        'area',                  # 9. Площадь (1 = маленькая, 2 = большая)
        'area_largest_spot',     # 10.Площадь наибольшего пятна (1 = <= 5, 2 => 5)
        'common_flares',         # 11.обычные вспышки
        'moderate_flares',       # 12.умеренные вспышки
        'severe_flares'          # 13.Сильные вспышки
    ]) 


f = open("./data/flare.data2","r")

data = [ a.strip().split(" ") for a in f][1:]
df = pd.DataFrame(np.array(data), columns=columns_name)  
df = df.drop('moderate_flares', axis='columns')
df = df.drop('severe_flares', axis='columns')
```

```python
def binarization_data_frame(df, column, array):
    for i in array:
        df[column + '_' + str(i)] = df[df[column] == i][column]
        df[column + '_' + str(i)][df[column + '_' + str(i)] == i] = 1
        df[column + '_' + str(i)] = df[column + '_' + str(i)].fillna(value=2)
    
    df = df.drop(column, axis='columns')
    return df  
```

```python
mod_zurix         = ['A', 'B', 'C', 'D', 'E', 'F', 'H']
largest_spot      = ['X', 'R', 'S', 'A', 'H', 'K']
spot_distribution = ['X', 'O', 'I', 'C']
evolution         = ['1', '2', '3']
activity_last_24  = ['1', '2', '3']
```

```python
df = binarization_data_frame(df, 'mod_zurix', mod_zurix)
df = binarization_data_frame(df, 'largest_spot', largest_spot)
df = binarization_data_frame(df, 'spot_distribution', spot_distribution)
df = binarization_data_frame(df, 'evolution', evolution)
df = binarization_data_frame(df, 'activity_last_24', activity_last_24)
```

```python
df['common_flares'] = df['common_flares'].replace(['2', '3', '4', '5', '6', '7', '8'], '1')
df['common_flares'] = df['common_flares'].replace(['0'], '2')
```
